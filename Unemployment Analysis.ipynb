```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "browser"
```

```python
# Read data
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")
print(data.head())
```

```python
# checking for the NULL values
```

```python
print(data.isnull().sum())
```

```python
# Rename Columns
```

```python
data.columns =["states", "Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region","Longitude","Latitude",]
```

```python
print(data)
```

```python
# corelation btw the features of dataset
```

```python
numeric_data = data.select_dtypes(include=['number'])
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Visualize Estimated Numbers Of Employes.
```

```python
data.columns =["states", "Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region","Longitude","Latitude"]
plt.title("Unemployment Rate")
sns.histplot(x="Estimated Employed",hue="Region",data=data)
plt.show()
```

```python
# Estimated Unemployment Rate.
```

```python
plt.figure(figsize=(12,10))
plt.title("Unemployment Rate")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=data)
plt.show()
```

```python
# Dashboard to Analyze Unemployment Rate of Each State by Region by usuing Sunburst plot.
```

```python
unemployment = data[["states","Estimated Unemployment Rate","Region"]]
figure = px.sunburst(unemployment,path=["Region","states"],
                    values="Estimated Unemployment Rate",
width =700,height=700, color_continuous_scale="RdYlGn",
title="Unemployment Rate")
figure.show()
```
